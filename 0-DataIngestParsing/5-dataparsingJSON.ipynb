```python
## Meethod 1- Using JSONLoader with jq_schema
from langchain_community.document_loaders import JSONLoader

employee_loader = JSONLoader(
    file_path="data/json_files/company_data.json",
    jq_schema=".employees[]",
    text_content=False
)

employees_docs = employee_loader.load()

print(f"length: {len(employees_docs)}")
print(f"content preview: {employees_docs[0].page_content[:200]}")
print(f"metadata: {employees_docs[0].metadata}")
```
